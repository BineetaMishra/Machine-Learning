```python
# Importing the library
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset

dataset = pd.read_csv('C:\\Users\\ebineet\\Documents\\Machine Learning\\Machine+Learning+A-Z+(Codes+and+Datasets)\\Machine Learning A-Z (Codes and Datasets)\\Part 2 - Regression\\Section 6 - Polynomial Regression\\Python\\Position_Salaries.csv')
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
#Training the dataset

# Training the linear regression model on the whole dataset
# fit method is a training method that will train the dataset

from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x, y)
```

```python
#Training the Plynomial regression model on the whole dataset

from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
x_poly = poly_reg.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly,y)
```

```python
#Visualizing
#Visualising the Linear Regression result

plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x), color='blue')
plt.title('True vs Bluff(Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
#Visualising the Ploynomial Regression result

plt.scatter(x,y,color='red')
plt.plot(x,lin_reg_2.predict(poly_reg.fit_transform(x)), color='blue')
plt.title('True vs Bluff(Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Ploynomial  Regression result Retrain with HIGHER DEGREE
#Retrain with a degree of 4

from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly,y)

#Visualising the Ploynomial Regression result
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg_2.predict(poly_reg.fit_transform(x)), color='blue')
plt.title('True vs Bluff(Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Ploynomial  Regression result (for higher resolution and smoother curve) 
# Increasing the ploting point from 1,2,3, --> (0.1,0.2,0.3,0.4,...)(x_grid = np.arange(min(x), max(x), 0.1))

x_grid = np.arange(min(x), max(x), 0.1)
x_grid = x_grid.reshape(len(x_grid), 1)
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg_2.predict(poly_reg.fit_transform(x)), color='blue')
plt.title('True vs Bluff(Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Predicting the result
#Predicting a new result with Linear Regression

lin_reg.predict([[6.5]])
```

```python
#Predicting a new result with Polynomial Regression 
#poly_reg.fit_transform because we want it in the polynomial form (x, x2, x3...)
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
